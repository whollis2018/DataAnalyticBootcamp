```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
#sns.set_style('darkgrid') #my default is "whitegrid"
```

```python
Location = "datasets/diamonds.csv"
df = pd.read_csv(Location)

df.head(200)
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols('carat ~ depth + table + price', data=df).fit()
```

```python
result.summary()
```

```python
#set coefficient to 0
result = smf.ols('carat ~ depth + table + price', data=df).fit()
result.summary()
```

```python
#make a scatterplot
sns.lmplot(x='carat', y='price', data=df)
```

```python
#boxplot showing stats for each characteristic
modeldf = df.drop(['Unnamed: 0','price'],axis=1)
sns.boxplot(data=modeldf)
```

```python
#change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of attack power for each type
sns.violinplot(x='color', y='depth', data=df)
```

```python
#swarm plot
df2 = df.head(107)
plt.subplots(figsize=(10,6))
sns.swarmplot(x='color', y='depth', data=df2)
```

```python
#combine a violin plot with a swarm plot
plt.figure(figsize=(13,7))
sns.violinplot(x='color',
               y='depth',
               data=df2,
               inner=None) #removes bars inside violin plot
             
sns.swarmplot(x='color',
              y='depth',
              data=df2,
              color='k', #make points black
              alpha=0.7) #slightly transparent

plt.title('Diamond Characteristics')
```

```python
#swarm plot of Stats from melted dataframe
sns.swarmplot(x='color', y='depth', 
              data=df2)
```
